```python
#    SVM        ,      ?
#  x, y  .   .
# 1) K(x, y) = xT y + b;
# 2) K(x, y) = (axT y + b)d
# 3) K(x, y) = tanh(axT+b)
# 4) K(x, y) = exp(||x-y||)
#      .

import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_circles
import warnings
warnings.filterwarnings('ignore')

data = make_circles(n_samples=300, noise=0.03, random_state=1)
X = data[:][0]
y = data[:][1]

plt.figure(figsize=(5,5))
plt.title("")
plt.scatter(X[:,0], X[:,1], c=y, s = 60)
plt.xlabel(" 0")
plt.ylabel(" 1")

from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)

for i in  ['linear', 'poly', 'rbf', 'sigmoid']:
    svm = SVC(random_state=0, kernel=i).fit(X_train,y_train)
    print("SVC.score train: {:.2f}% | SVC.score test: {:.2f}% | {}\n".format(svm.score(X_train, y_train)*100, svm.score(X_test, y_test)*100, i))
```
